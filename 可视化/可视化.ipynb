```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from sklearn.linear_model import LinearRegression


# 获取增长率
def get_zzl(value):
    values = [1.47]
    for i in value:
        values.append(i)
    zzl = []
    for i in range(1,len(values)):
        num = (values[i] - values[i-1]) / values[i-1]
        num = round(num*100,2)
        zzl.append(num)
    return zzl


def scgm():
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False

    data = pd.read_csv("./所有数据集/市场规模.csv",encoding='gbk')
    print(data)

    year = list(data['年份'])[1::]
    value = list((data['市场规模']))[1::]
    print("年份：",year)
    print("市场规模：",value)
    zzl = get_zzl(value)
    print(zzl)


    figsize = 16,9
    figure, ax1 = plt.subplots(figsize=figsize, dpi=100)

    # 柱状图
    ax1.bar(year, value, 0.35, color='#1F77B4', label='物联网行业市场经济规模(万亿元)')
    for i in range(len(year)):
        ax1.text(year[i], value[i], str(value[i]), ha='center', va='bottom', fontsize=20, fontweight='bold')
    ax1.set_xlabel("年份", fontsize=20)
    ax1.set_ylabel("物联网行业市场规模年度数据", fontsize=20)
    ax1_y_position = np.arange(0, 5.1, 0.5)
    ax1.set_yticks(ax1_y_position)
    ax1.set_yticklabels(ax1_y_position,fontsize=20)
    plt.xticks(fontsize=20, rotation=10)
    # 网格虚线
    plt.gca().yaxis.set_major_locator(MultipleLocator(1))
    plt.gca().yaxis.set_minor_locator(MultipleLocator(0.2))
    xu_line = [i/10 for i in range(0,50,2) if i not in [10,20,30,40]]
    for i  in xu_line[1::]:
        plt.axhline(y=i,linestyle='--',color='#B2B2B2',alpha=0.5) 
    for i in [1,2,3,4,5]:
        plt.axhline(y=i,linestyle='-',color='#B2B2B2',alpha=0.5)

   
    # 折线图
    ax2 = plt.twinx()
    ax2.plot(year, zzl, color='#FE7F0e', marker='o', label='增长率(%)')
    # 在每个点的旁边显示数据
    for i in range(len(year)):
        ax2.text(year[i], zzl[i], str(zzl[i])+'%', ha='right', va='bottom', fontsize=20, color='#FE7F0e',fontweight='bold')

    ax2.set_ylabel("物联网行业市场规模增长率", fontsize=16)
    ax2.spines['right'].set_position(('outward', 0))
    ax2_y_position = np.arange(0, 31, 5)
    y_label = [str(i)+'%' for i in ax2_y_position]
    ax2.set_yticks(ax2_y_position)
    ax2.set_yticklabels(y_label,fontsize=20)

    ax1.set_title("2019年到2024年中国物联网行业市场经济规模变化情况", fontsize=25)
    ax1.legend(loc='upper left', fontsize=17)
    ax2.legend(loc='upper right', fontsize=17)
    plt.show()    


if __name__ == '__main__':
    scgm()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from sklearn.linear_model import LinearRegression
import jieba
from sklearn.feature_extraction.text import CountVectorizer


def cut_word(text):
    # print(" ".join(list(jieba.cut(text))))
    return " ".join(list(jieba.cut(text)))


def get_feature():
    data_BossZP = pd.read_csv("./所有数据集/BossZP.csv",encoding='gbk')
    data_QCWY = pd.read_csv("./所有数据集/QCWY_new.csv",encoding='gbk')
    type1 = list(data_BossZP['BrandIndustry']) + list(data_QCWY['Industry'])
    type2 = list(data_QCWY['Industry'])
    print(type1)
    print(type2)
    for i in range(len(type2)):
        type2[i] = type2[i][2:-2]
    text = type1 + type2
    text_new = []
    for sentence in text:
        text_new.append(cut_word(sentence))
    transform = CountVectorizer()
    data = transform.fit_transform(text_new)
    data_list = data.toarray()
    column_sum = np.sum(data_list, axis=0)
    feature_names = transform.get_feature_names_out()
    return dict(zip(list(feature_names), list(column_sum)))
    
    
def butwlxxq():
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False

    data = get_feature()
    sorted_data = dict(sorted(data.items(), key=lambda item: item[1], reverse=True))
    type_name = list(sorted_data.keys())
    type_data = list(sorted_data.values())
    print(type_name)
    print(type_data)
    sum_data = sum(type_data)
    # 饼图
    figsize = 16,9
    figure, ax1 = plt.subplots(figsize=figsize, dpi=100)
    ax1.pie(type_data[0:20], labels=type_name[0:20], autopct='%1.1f%%', shadow=False, startangle=140)
    plt.axis('equal')
    plt.legend()
    plt.title('物联网行业相关岗位领域分布图', fontsize=20)


if __name__ == '__main__':
    # get_feature()
    butwlxxq()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.cm as cm
import seaborn as sns
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from sklearn.linear_model import LinearRegression
import jieba
from sklearn.feature_extraction.text import CountVectorizer
import ast
from collections import Counter
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.charts import Timeline
from pyecharts.charts import Map
import math


def deal_with_BossZP_salary_data(salary):
    if salary[-1]=='天':
        salary = salary[0:-3]
        salary_min = int(salary.split('-')[0])
        salary_max = int(salary.split('-')[1])
        salary = (salary_min+salary_max)/2 *30
    elif salary[-1]=='K':
        salary = salary[0:-1]
        salary_min = int(salary.split('-')[0])
        salary_max = int(salary.split('-')[1])
        salary = (salary_min+salary_max)/2 *1000
    elif salary[-1]=='薪':
        salary = salary[0:-1]
        # salary_month = int(salary.split('·')[1])
        salary = salary.split('·')[0][0:-1]
        salary_min = int(salary.split('-')[0])
        salary_max = int(salary.split('-')[1])
        salary = (salary_min+salary_max)/2 *1000
    elif salary[-1]=='万':
        salary = salary[0:-1]
    elif salary[-1]=='月':
        salary = salary[0:-3]
        salary_min = int(salary.split('-')[0])
        salary_max = int(salary.split('-')[1])
        salary = (salary_min+salary_max)/2
    elif salary[-1]=='周':
        salary = salary[0:-3]
        salary_min = int(salary.split('-')[0])
        salary_max = int(salary.split('-')[1])
        salary = (salary_min+salary_max)/2 *4
    return salary

def deal_with_QCWY_salary_data(salary):
    salary_new = None
    if salary[-1]=='薪':
        salary = salary.split('·')[0]
        if salary[-1]=='千':
            salary_min = int(salary.split('-')[0])
            salary_max = int(salary.split('-')[1][0:-1])
            salary_new = (salary_min*1000+salary_max*1000)/2
            # print(salary_new)
        elif salary[-1]=='万':
            salary_min = salary.split('-')[0]
            salary_max = salary.split('-')[1]
            if salary_min[-1]=='千':
                salary_min = float(salary_min[0:-1])
                salary_max = float(salary_max[0:-1])
                salary_new = (salary_min*1000+salary_max*10000)/2
            else:
                salary_min = float(salary_min)
                salary_max = float(salary_max[0:-1])
                salary_new = (salary_min*10000+salary_max*10000)/2
    if salary[-1]=='万':
        salary_min = salary.split('-')[0]
        salary_max = salary.split('-')[1]
        if salary_min[-1]=='千':
            salary_min = float(salary_min[0:-1])
            salary_max = float(salary_max[0:-1])
            salary_new = (salary_min*1000+salary_max*10000)/2
        else:
            salary_min = float(salary_min)
            salary_max = float(salary_max[0:-1])
            salary_new = (salary_min*10000+salary_max*10000)/2
    elif salary[-1]=='千':
        salary_min = float(salary.split('-')[0])
        salary_max = float(salary.split('-')[1][0:-1])
        salary_new = (salary_min*1000+salary_max*1000)/2
        # print(salary_min,salary_max)
    elif salary[-1]=='年':
        salary = salary[0:-3]
        salary_min = int(salary.split('-')[0])
        salary_max = int(salary.split('-')[1])
        middle_salary = (salary_min+salary_max)/2
        salary_new = middle_salary/12*10000
        # print(salary_min,salary_max)
    # print(salary)
    return salary_new

def get_BossZP_salary_data():
    data = pd.read_csv("./所有数据集/BossZP.csv",encoding='gbk')
    salary = list(data['Salary Description'])
    salary_data = []
    for i in salary:
        salary_data.append(deal_with_BossZP_salary_data(i))
    # print(salary_data)
    return salary_data

def get_QCWY_salary_data():
    data = pd.read_csv("./所有数据集/QCWY_new.csv",encoding='gbk')
    salary = list(data['Job Salary'])
    # print(salary)
    salary_data = []
    for i in salary:
        # print(i)
        salary_data.append(deal_with_QCWY_salary_data(i))
    # print(salary_data)
    return salary_data

def get_BossZP_degree_data():
    data = pd.read_csv("./所有数据集/BossZP.csv",encoding='gbk')
    degree = list(data['Job Degree'])
    return degree

def get_QCWY_degree_data():
    data = pd.read_csv("./所有数据集/QCWY_new.csv",encoding='gbk')
    degree = list(data['Job Degree'])
    return degree
    
def get_BossZP_experience_data():
    data = pd.read_csv("./所有数据集/BossZP.csv",encoding='gbk')
    experience = list(data['Job Experience'])
    return experience

def get_QCWY_experience_data():
    data = pd.read_csv("./所有数据集/QCWY_new.csv",encoding='gbk')
    experience = list(data['Job Year'])
    return experience
    

def BossZP_salary_degree():
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    salary = get_BossZP_salary_data()
    s = pd.Series(salary)
    median_value = s.median()
    degree = get_BossZP_degree_data()
    order_degree = ['学历不限','大专','本科']
    sns.catplot(x=degree,
                y=salary,
                kind='strip',
                order=order_degree)
    plt.axhline(y=median_value, color='r', linestyle='--') 
    plt.text(-0.5, median_value, f'{median_value}', color='red',ha='right')
    plt.xlabel('学历要求')
    plt.ylabel('薪资')
    plt.title('物联网行业就业薪资与学历要求的关系（Boss直聘）')


def BossZP_salary_experience():
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    salary = get_BossZP_salary_data()
    s = pd.Series(salary)
    median_value = s.median()
    experience = get_BossZP_experience_data()
    order_experience = ['在校/应届','1年以内','1-3年','3-5年',' 5-10年','10年以上','经验不限']
    sns.catplot(x=experience,
                y=salary,
                kind='strip',
                order=order_experience)
    plt.axhline(y=median_value, color='r', linestyle='--') 
    plt.text(-0.5, median_value, f'{median_value}', color='red',ha='right')
    plt.xlabel('经验要求')
    plt.ylabel('薪资')
    plt.title('物联网行业就业薪资与经验要求的关系（Boss直聘）')


def BossZP_salary_degree_experience():
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    salary = get_BossZP_salary_data()
    s = pd.Series(salary)
    median_value = s.median()
    degree = get_BossZP_degree_data()
    experience = get_BossZP_experience_data()
    order_experience = ['在校/应届','1年以内','1-3年','3-5年',' 5-10年','10年以上','经验不限']
    sns.catplot(x=experience,
                y=salary,
                hue=degree,
                kind='strip',
                order=order_experience)
    plt.axhline(y=median_value, color='r', linestyle='--') 
    plt.text(-0.5, median_value, f'{median_value}', color='red',ha='right')
    # plt.xlabel('学历要求')
    plt.xlabel('经验要求')
    plt.ylabel('薪资')
    plt.title('物联网行业就业薪资与学历要求和经验要求的关系（Boss直聘）')


def QCWY_salary_degree():
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    salary = get_QCWY_salary_data()
    # print(salary)
    s = pd.Series(salary)
    median_value = s.median()
    degree = get_QCWY_degree_data()
    order_degree = ['初中及以下','中技/中专','高中','大专','本科','硕士','博士']
    sns.catplot(x=degree,
                y=salary,
                kind='strip',
                order=order_degree)
    plt.axhline(y=median_value, color='r', linestyle='--') 
    plt.text(-0.5, median_value, f'{median_value}', color='red',ha='right')
    plt.xlabel('学历要求')
    plt.ylabel('薪资')
    plt.title('物联网行业就业薪资与学历要求的关系（前程无忧）')


def QCWY_salary_experience():
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    salary = get_QCWY_salary_data()
    s = pd.Series(salary)
    median_value = s.median()
    experience = get_QCWY_experience_data()
    order_experience = ['无需经验','1年','1-2年','2年','1-3年','2-3年','3-4年',
                        '3-5年','3-6年','3-9年','4-6年','5-7年','5-8年','5-10年',
                        '8-9年','1年及以上','2年及以上','3年及以上','4年及以上',
                        '5年及以上','6年及以上','8年及以上','10年及以上']
    sns.catplot(x=experience,
                y=salary,
                kind='strip',
                order=order_experience)
    plt.axhline(y=median_value, color='r', linestyle='--') 
    plt.text(-0.5, median_value, f'{median_value}', color='red',ha='right')
    plt.xlabel('经验要求')
    plt.ylabel('薪资')
    plt.xticks(rotation=90)
    plt.title('物联网行业就业薪资与经验要求的关系（前程无忧）')


def QCWY_salary_degree_experience():
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    salary = get_QCWY_salary_data()
    s = pd.Series(salary)
    median_value = s.median()
    degree = get_QCWY_degree_data()
    experience = get_QCWY_experience_data()
    order_experience = ['无需经验','1年','1-2年','2年','1-3年','2-3年','3-4年',
                        '3-5年','3-6年','3-9年','4-6年','5-7年','5-8年','5-10年',
                        '8-9年','1年及以上','2年及以上','3年及以上','4年及以上',
                        '5年及以上','6年及以上','8年及以上','10年及以上']
    sns.catplot(x=experience,
                y=salary,
                hue=degree,
                kind='strip',
                order=order_experience)
    plt.xticks(rotation=90)
    plt.axhline(y=median_value, color='r', linestyle='--') 
    plt.text(-0.5, median_value, f'{median_value}', color='red',ha='right')
    # plt.xlabel('学历要求')
    plt.xlabel('经验要求')
    plt.ylabel('薪资')
    plt.title('物联网行业就业薪资与学历要求和经验要求的关系（前程无忧）')




if __name__ == '__main__':
    BossZP_salary_degree()
    BossZP_salary_experience()
    BossZP_salary_degree_experience()
    QCWY_salary_degree()
    QCWY_salary_experience()
    QCWY_salary_degree_experience()
```
